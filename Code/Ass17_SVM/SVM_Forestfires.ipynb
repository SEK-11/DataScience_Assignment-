```python
def outlier(data,colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)

    iqr = q3 - q1

    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)

    return lower, upper
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
x = df.drop(columns=["size_category"])
y = df["size_category"]
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
svc = SVC()
svc.fit(xtrain, ytrain)
ypred = svc.predict(xtest)
ypred
```

```python
svc.score(xtrain, ytrain), svc.score(xtest, ytest)
```

```python
param_grid = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [0, 2, 3, 4]
}
```

```python
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2, scoring='accuracy')
grid_search.fit(xtrain, ytrain)
```

```python
best_params = grid_search.best_params_
best_params
```

```python
svc = SVC(degree = 0, kernel = 'linear')
svc.fit(xtrain, ytrain)
ypred = svc.predict(xtest)
ypred
```
