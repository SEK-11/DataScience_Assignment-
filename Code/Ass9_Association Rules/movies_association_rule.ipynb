```python
#Import Libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/SEK-11/dataset_excelR/main/DATA/my_movies.csv")
```

```python
df
```

```python
df.drop(columns=['Sixth Sense','Gladiator','LOTR1','Harry Potter1','Patriot','LOTR2','Harry Potter2','LOTR','Braveheart','Green Mile'])
```

```python
data = pd.get_dummies(df)
data
```

```python
scores = apriori(data,min_support=0.2,use_colnames=True)
```

```python
scores
```

```python
association_rules(scores)
```
