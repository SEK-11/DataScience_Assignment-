```python
#Import Libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/SEK-11/dataset_excelR/main/DATA/book.csv")
```

```python
df
```

```python
scores = apriori(df,min_support=0.2,use_colnames=True)
```

```python
scores
```

```python
association_rules(scores)
```
